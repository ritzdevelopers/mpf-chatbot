```python
from langchain.schema import Document
```

```python
import json
with open("property_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
with open("cities.json", "r", encoding="utf-8") as f:
    cities = json.load(f)
```

```python
from langchain.schema import Document

documents = []

# =========================
# CITY DOCUMENTS
# =========================

for city in cities:

    metadata = {
        "type": "city",
        "metaTitle": city.get("metaTitle"),
        "metaKeywords": city.get("metaKeywords"),
        "metaDescription": city.get("metaDescription"),
        "cityName": city.get("cityName"),
        "stateName": city.get("stateName"),
        "countryName": city.get("countryName")
    }

    content = str(city.get("cityDescription", "")).strip()

    if not content:
        continue

    doc = Document(
        page_content=content,
        metadata=metadata
    )

    documents.append(doc)


# =========================
# PROPERTY DOCUMENTS
# =========================

extract_cols = [
    "metaTitle",
    "metaKeyword",
    "projectName",
    "projectLocality",
    "projectConfiguration",
    "state",
    "city",
    "country",
    "propertyTypeName"
]

for item in data:

    metadata = {
        col: item.get(col)
        for col in extract_cols
    }

    metadata["type"] = "property"
    metadata["images"] = item.get("mobileImages")
    metadata["metaDescription"] = item.get("metaDescription")

    content = str(item.get("final_content", "")).strip()

    if not content:
        continue

    doc = Document(
        page_content=content,
        metadata=metadata
    )

    documents.append(doc)

```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
chunks = splitter.split_documents(documents)
```
